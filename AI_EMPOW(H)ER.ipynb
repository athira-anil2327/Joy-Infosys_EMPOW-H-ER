# 🎨 **Step 2: Write the Streamlit Application Script**
This cell uses the `%%writefile` command to generate the core `app.py` script. It initializes the custom velvet maroon CSS theme, maps out the structured mock response engine for visual/text modes, and lays out the interactive UI components.

```python
%%writefile app.py
import streamlit as st
import time

MOCK_RESPONSES = {
    "🌟 Explain it simply!": (
        "🌟 **Hey there! Let's break down {topic} together!** 🚀\n\n"
        "Think of **{topic}** like organizing a messy shelf. Instead of looking at every single item "
        "one by one (which takes forever!), we use a smart system to find exactly what we need in seconds!\n\n"
        "✨ **Key takeaway:** It saves our computer a massive amount of brainpower. "
        "You are doing an amazing job tackling this concept today! Keep shining! 🎉🧠"
    ),
    "🍿 Show me a fun analogy!": (
        "🍿 **Story time! Imagine you are looking for a specific page in a massive textbook...** 📚\n\n"
        "If you were using a basic approach for **{topic}**, you might flip every single page from the beginning. "
        "But instead, you open it right to the middle! If the chapter you need is further ahead, you completely "
        "ignore the first half of the book. \n\n"
        "💃 That is exactly how **{topic}** works in programming—it eliminates half the work at every single step! "
        "Super elegant, right? You're mastering this! 🌟"
    ),
    "🧠 Challenge me with a friendly quiz!": (
        "🧠 **Time for a fun, stress-free check-in on {topic}! You've got this!** 🚀\n\n"
        "**Q1:** What is the primary goal of implementing **{topic}**?\n"
        "👉 *A) To make the code look prettier*\n"
        "👉 *B) To find data much faster and save processing time*\n"
        "👉 *C) To delete old files*\n\n"
        "**Q2:** True or False: Mastering **{topic}** makes you an official coding superstar.\n"
        "👉 *A) True (Absolutely!)*\n"
        "👉 *B) False*\n\n"
        "--- \n"
        "🎁 *Scroll down for the answers...* \n\n\n\n"
        "🎉 **Answers:** 1-B, 2-A! Incredibly well done! Celebrate your progress! 🌟"
    )
}

st.set_page_config(page_title="Joy | Your AI Learning Buddy", page_icon="✨", layout="wide")

st.markdown("""
    <style>
    [data-testid="stAppViewContainer"] {
        background: linear-gradient(135deg, #3C0A11 0%, #1A0205 100%) !important;
        color: #FFF5F5 !important;
    }
    [data-testid="stHeader"] {
        background: transparent !important;
    }
    [data-testid="stSidebar"] {
        background-color: #240509 !important;
        border-right: 2px solid #5C131D;
    }
    [data-testid="stSidebar"] .stMarkdown p, [data-testid="stSidebar"] label {
        color: #FFD2D7 !important;
    }
    h1, h3 {
        color: #FFF5F5 !important;
        text-shadow: 0px 0px 8px rgba(255, 100, 120, 0.3);
    }
    .stButton>button {
        background: linear-gradient(90deg, #A82030, #5C131D) !important;
        color: #FFFFFF !important;
        border: 1px solid #FF6B7D !important;
        border-radius: 20px !important;
        font-weight: bold !important;
        transition: 0.3s ease-in-out;
    }
    .stButton>button:hover {
        background: linear-gradient(90deg, #D43F51, #7E1E2A) !important;
        box-shadow: 0px 0px 12px #FF6B7D !important;
        transform: scale(1.02);
    }
    div[data-baseweb="input"] input {
        color: #FFFFFF !important;
        background-color: #4A1017 !important;
        border-radius: 8px !important;
    }
    .stAlert {
        background-color: #3D0B12 !important;
        border: 1px solid #8B1F2D !important;
        color: #FFE5E8 !important;
    }
    </style>
""", unsafe_allow_html=True)

st.sidebar.markdown("## ⚙️ Configuration Panel")
action = st.sidebar.radio(
    "What should we do first?",
    ["🌟 Explain it simply!", "🍿 Show me a fun analogy!", "🧠 Challenge me with a friendly quiz!"]
)

show_visuals = st.sidebar.checkbox("🖼️ Include visual reference charts/diagrams", value=True)

st.markdown("# ✨ Meet Joy: Your AI Learning Buddy! ✨")
st.markdown("### *Let's master algorithms together with zero stress and lots of fun! 💖*")
st.write("---")

topic = st.text_input("🎨 What beautiful concept are we learning today?", value="Binary Search")

if st.button("🚀 Let's Go!"):
    if not topic:
        st.warning("Oops! Don't forget to type a topic first! 🥰")
    else:
        with st.spinner("🤖 Joy is brewing up some magic... ✨"):
            time.sleep(0.8)

            if show_visuals:
                col1, col2 = st.columns([2, 1])

                with col1:
                    st.success(f"🎉 Joy's breakdown for {topic}:")
                    response_text = MOCK_RESPONSES[action].format(topic=topic)
                    st.markdown(response_text)

                with col2:
                    st.info("🖼️ Algorithm Visualizer")
                    st.image(
                        "https://images.unsplash.com/photo-1515879218367-8466d910aaa4?w=500",
                        caption=f"Visual structural flow diagram representing: {topic}",
                        use_container_width=True
                    )
            else:
                st.success("🎉 Here is what Joy prepared for you:")
                response_text = MOCK_RESPONSES[action].format(topic=topic)
                st.markdown(response_text)
```
